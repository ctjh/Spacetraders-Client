```python
class System():
    def __init__(self,symbol):
        self.symbol = symbol
    def getWaypointDetails(self,waypoint):
        response = requests.get(f'https://api.spacetraders.io/v2/systems/{self.symbol}/waypoints/{waypoint}')
        return response.json()['data']
    def findWaypointByType(self, _type):
        response = requests.get(f'https://api.spacetraders.io/v2/systems/{self.symbol}/waypoints?type={_type}')
        return response.json()['data']
    def findWaypointByTrait(self, trait):
        response = requests.get(f'https://api.spacetraders.io/v2/systems/{self.symbol}/waypoints?traits={trait}')       
        return response.json()['data']
    def marketplaceDetails(self,marketplace):
        response = requests.get(f'https://api.spacetraders.io/v2/systems/{self.symbol}/waypoints/{marketplace}/market')
        return response.json()['data']
    def shipyardDetails(self,shipyard):
        response = requests.get(f'https://api.spacetraders.io/v2/systems/{self.symbol}/waypoints/{shipyard}/shipyard')
        return response.json()['data']        
    
    def findNearestCoordinateByTrait(self,ship,trait): #Ship Object and list of coordinates
        details = shipOne.getNavStatus()
        shipX, shipY =  details['route']['destination']['x'], details['route']['destination']['y']
        start = [shipX,shipY]
        coordinates = []
        locations = self.findWaypointByTrait(trait)
        for location in locations:
            coordinates.append([location['x'], location['y'],location['symbol']])
        def euclidean_distance(point1, point2):
            return math.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)
        nearest_coordinate = None
        min_distance = float('inf')
        for coord in coordinates:
            distance = euclidean_distance(start, coord)
            if distance < min_distance:
                min_distance = distance
                nearest_coordinate = coord
        return nearest_coordinate
system = System('X1-RR47')
# system.getWaypointDetails('X1-RR47-A2')
# system.findWaypointByTrait('COMMON_METAL_DEPOSITS')
# system.shipyardDetails('X1-RR47-A2')
# print(system.findNearestCoordinateByTrait(shipOne,'COMMON_METAL_DEPOSITS'))
# print(system.findNearestCoordinateByTrait(shipOne,'MARKETPLACE'))
# system.marketplaceDetails('X1-RR47-A2')
# print(system.findNearestCoordinateByTrait(shipOne,'REFUEL_STATION'))
```
